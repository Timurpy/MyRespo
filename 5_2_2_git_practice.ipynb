Чтобы не путать публичную историю проекта, вместо неё лучше делать ещё коммит с обратными изменениями, например, вот так: `git revert HEAD~1`.

**Важно!**<br>
Для записи в удаленный репозиторий вас попросят идентифицировать себя. Логин и пароль при этом не подойдет: в августе 2021 года по требованиям безопасносности процедура доступа <a href="https://github.blog/2020-12-15-token-authentication-requirements-for-git-operations/">была изменена</a>. 

Самое простое что можно сделать -- настроить себе временный токен доступа в веб-интерфесе (<a href="https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/creating-a-personal-access-token">Инструкция</a>). Скопируйте этот токен после генерации, сохраните его и используйте вместо логина (с пустым паролем).

Если вы пользуетесь гитом профессионально, то скорее всего вам потребуется другое решение -- генерация SSH ключа на вашей машине/на сервере и его сохранение в ssh-клиенте и в веб-интерфейсе GitHub. <a href="https://docs.github.com/en/authentication/connecting-to-github-with-ssh">Здесь</a> можно почитать подробнее.
